```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = np.array([[10,15,20,25, 30]]).T
y = np.array([[10,30,50,51, 52]]).T
```

```python
linear = LinearRegression()
linear.fit(X,y)
print("Coef ", linear.coef_)
print("Intercept ", linear.intercept_)
print("R^2 ", linear.score(X,y))
print()
pred = linear.predict(X)
#X_test = np.array([[-10,10,15,20,25, 30,50]]).T
#print(linear.predict(X_test))
print("MSE Linear = ", mean_squared_error(y,pred))

plt.plot(X,y,'o')
plt.plot(X,pred,'r-',label="linear regression")
#plt.axis('equal')
plt.legend() 
plt.show()
```

```python
poly_reg=PolynomialFeatures(degree=2)
#print("X ",X)

X_poly = poly_reg.fit_transform(X)
#print("X_poly ", X_poly)

lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,y)
y_pred = lin_reg2.predict(X_poly)

plt.scatter(X,y)
plt.plot(X,y_pred,'g-',label="Polynomial regression")
#plt.axis('equal')
plt.legend() 
plt.show()

print("MSE polynomial = ", mean_squared_error(y,y_pred))
```

```python
poly_reg=PolynomialFeatures(degree=4)
#print("X ",X)

X_poly = poly_reg.fit_transform(X)
#print("X_poly ", X_poly)

lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,y)
y_pred = lin_reg2.predict(X_poly)
#print(lin_reg2.predict(poly_reg.fit_transform(np.array([[12]]).T)))

plt.scatter(X,y)
plt.plot(X,y_pred,'g-',label="Polynomial regression")
#plt.axis('equal')
plt.legend() 
plt.show()

print("MSE polynomial = ", mean_squared_error(y,y_pred))
```

```python
from sklearn.linear_model import Ridge
import numpy as np
n_samples, n_features = 10, 5
rng = np.random.RandomState(0)
y = rng.randn(n_samples)
X = rng.randn(n_samples, n_features)

n_alphas = 10
alphas = np.linspace(0, 10, n_alphas)
coefs = []
for a in alphas:
    clf = Ridge(alpha=a)
    clf.fit(X, y)
    print("alpha = ",a, " coef = ", clf.coef_)
#plt.scatter(X,y)
#plt.scatter(X,clf.predict(X))
```

```python
# Author: Fabian Pedregosa -- <fabian.pedregosa@inria.fr>
# License: BSD 3 clause

print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# X is the 10x10 Hilbert matrix
X = 1. / (np.arange(1, 11) + np.arange(0, 10)[:, np.newaxis])
y = np.ones(10)
# print(y)
# #############################################################################
# Compute paths

n_alphas = 100
alphas = np.logspace(-10, -2, n_alphas)
#print(alphas)
coefs = []
for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)
    #print(coefs)

# #############################################################################
# Display results

ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```
