```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error

X = np.array([[60,80,100,100]]).T
y = np.array([[170,160,180,200]]).T

ridge = Ridge(alpha=1)
ridge.fit(X, y)
print("ridge.coef_", ridge.coef_)
print("ridge.intercept_", ridge.intercept_)
print("ridge R^2",ridge.score(X,y))
print()
linear = LinearRegression()
linear.fit(X,y)
print("ols.coef_", linear.coef_)
print("ols.intercept_", linear.intercept_)
print("ols R^2", linear.score(X,y))
print()

yr_predict = ridge.predict(X)
yl_predict = linear.predict(X)
print("MSE Linear = ", mean_squared_error(y,yl_predict))
print("MSE ridge = ", mean_squared_error(y,yr_predict))

plt.plot(X,y,'o')
plt.plot(X,yr_predict,'r-',label="ridge")
plt.plot(X,yl_predict,'g-',label="linear")
plt.legend() 
plt.show()
```

```python
from sklearn import linear_model
lasso = linear_model.Lasso(alpha=1)
lasso.fit(X,y)
print("lasso.coef_", lasso.coef_)
print("lasso.intercept_", lasso.intercept_)
print("lasso R^2", lasso.score(X,y))

ys_predict = lasso.predict(X)
print("MSE Linear = ", mean_squared_error(y,yl_predict))
print("MSE ridge = ", mean_squared_error(y,yr_predict))
print("MSE lasso = ", mean_squared_error(y,ys_predict))

plt.plot(X,y,'o')
plt.plot(X,yr_predict,'r-',label="ridge")
plt.plot(X,yl_predict,'g-',label="linear")
plt.plot(X,ys_predict,'b-',label="lasso")
plt.legend() 
plt.show()
```
